# Assignment Operators
`=, +=, -=, *=, /=, %=, **=, //=`


```python
a=50
#used to work right to left
#var=data (equal to used to store data into a variable)
print(a)
```

```python
a,b=4,5 
print(a,b) 
```

###Above process is called Tuple Unpacking

```python
#NOTE- Equal to = first solves right side then stores those values to left side
#left=identifier, right=data 
a,b=4,5 #implies a=4 and b=5 
a,b=b**2,a**2 
print(a,b) #a=b**2=5**2=5^2=25 & b=a**2=4**2=4^2=16
#therefore a=25, b=16 
```

`swapping`

```python
x=10
y=50

t=x #temp=10
x=y #x=50
y=t #y=10

print(x,y) #swapping
```

```python
#Swap without using 3rd variable(Temp)
x=10
y=50

x=x+y #10+50=60

y=x-y #60-50=10

x=x-y #60-10=50

print(x,y)
```

```python
#another way easiest and simple
x=10
y=50

x, y=y, x

print(x,y)
```
