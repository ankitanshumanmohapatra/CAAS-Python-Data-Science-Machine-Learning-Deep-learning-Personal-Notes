````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
data = [
    ["A", 50, 60, 80],
    ["B", 56, 45, 87],
    ["C", 56, 45, 87],
    ["D", 58, 79, 34],
    ["E", 34, 76, 87]
]

df = pd.DataFrame(
    data,
    columns=["name", "maths", "chemistry", "physics"],
    index=[1001, 1010, 1007, 1313, 1111]                         #here coloumn indexes we assigned to dataframe
)
````

````python
df
````

````python
df.set_index('name') #"name" column set as inplace of "index"  (But this just creates a copy of the original data frame and original dataframe is secured)
# copy of DataFrame is which is created
# inplace=True -> change in original DataFrame
````

````python
df #as we can still original data frame is same we are not changing it
````

````python
df.reset_index() #reset function creates a new coloumn (index is created as a coloumn)
````

````python
df
````

````python
df.loc[1111, 'maths'] # 34
````

````python
df.at[1111, 'maths'] # 34
````

````python
df.iat[4, 1]
````

    loc -> label indexes (row labels, col labels)
    iloc -> default index
    
    
    at -> scaler value selection using label indexes
    iat -> scaler value selection using default indexes
    
    at and iat are way faster compare to loc, iloc
