##Method2 
**`To add only new key-value pairs in dictionary`**

    dict.setdefault(key, default=None) - if key is not there in dict then update key with default value, return value at key

here as key if already exists so it will not update it with given value

```python
#return value
value = info.setdefault('name', 'Sachin Yadav')
print(value)
#here as name key already exists so it will not update it with sachin
```
