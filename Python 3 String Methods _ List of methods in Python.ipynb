l.just means left justification with width 100 as x(window size) then **fillchar** means what character we will use to fill the blank spaces

```python
msg = "Hello World!"
width = 30

s1= msg.ljust(width)   #left justification of message
print(repr(s1), "Left Justification")       #returns a function representation of an object in python
s2= msg.center(width)  #center justification of message
print(repr(s2), "Right Justification")
s3=msg.rjust(width)         #right justification of message
print(repr(s3), "Centre Justification")
```

#sort()
####sort methods means arrainging data in ascending or descending order
####here in python we have a built-in function for sorting

```python
nums = [ 14, 56, 78, 56, 45, 34, 56, 78, 54, 89, 45, 23]
```

```python
#sort method only works for homogenous list
nums.sort() #reverse=True for descending
print(nums)
```

```python
nums = [ 14, 56, 78, 56, 45, 34, 56, 78, 54, 89, 45, 23]
nums.sort(reverse=True)
print(nums)
```
