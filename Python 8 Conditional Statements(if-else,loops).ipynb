 a year is consider a leap if it is divisible by 4 if it is not a century year
 if year is century year than if it is divsible by 400 it is leap year

    2016 -> leap year -> divisible by 4, not a century year

    1900 -> leap year -> divisble by 4, a century year, it is not divisible by 400

## Nested if-else
if and else and inside the another if else

````python
#1st check century year then directly leaf year or not then check if div by 4 then leaf
year = int(input("year: "))

if year % 100 == 0:
# year is century year
    if year % 400 == 0:
        print(f"{year} is a Leap Year")
    else:
        print(f"{year} is not a Leap Year")
else:
# year is not century year
    if year % 4 == 0:
        print(f"{year} is a Leap Year")
    else:
        print(f"{year} is Not a Leap Year")
````

````python
#record list traversing
i=0
while i<len(records):
  print(i, records[i])
  i+=1
````

````python
#from a Student record find a student is present or not
i = 0
while i < len(records):
    name = records[i][0] #record i 0th position gets name access
    per = sum(records[i][1:])/3 #ith report and sum of records/3
    print(f"{i+1}. {name:>10} = {per:.2f}") #percent till 2 decimal place so 2f
    i += 1
````

````python
name = input("student name: ").strip().lower()
i = 0
while i < len(records):
    st = records[i]
    if st[0] == name: 
        per = sum(st[1:])/3 
        print(f"{name} has scored {per:.2f}% marks.")
        break
    i += 1
else:
    print("Student Record Not Found!")
````

## Continue Statement

      continues with loop 
      below example instead of break use continue then get continues the iteration

image.png

````python
c = 1
while c <= 10:
    if c % 2 == 0:
        break
    print(c)
    c += 1
else:
    print("i hope all will give correct answer")
````

````python
c = 1
while c <= 10:
    if c % 2 == 0:
        continue
    print(c)
    c += 1
else:
    print("i hope all will give correct answer")
    
# ? #Infinite loop continue.....
````
